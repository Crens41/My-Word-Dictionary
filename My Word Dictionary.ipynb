```python
quiz = {
    "question1": {
        "question" : "What is the square root of 25?",
        "answer" : "625"
    },
    "question2" : {
        "question" : "Who was called the saviour of the world?",
        "answer" : "Jesus"
    },
    "question3" : {
        "question" : "How many days are there in a leap year?",
        "answer": "366"
    },
    "question4": {
        "question" : "What is the product of 8 and 10?",
        "answer": "80"
    },
    "question5" : {
        "question" : "What year did the Queen of England die?",
        "answer": "2022"
    }
}

score = 0
for key, value in quiz.items():
    print(value['question'])
    answer = input("insert your answer: ")
    
    if answer.lower() == value["answer"].lower():
        print("Your answer is correct")
        score = score + 1
        print("Your score is:" + str(score))
        print("")
        print("")
        print("")
        
    else:
        print("You are wrong")
        print("your score is: " + str(score))
        print("")
        print("")
        print("")
print("Your total score is: " + str(score) + "/5")
print("Your percentage is: " + str(int(score/5 * 100)) + "%")
```
